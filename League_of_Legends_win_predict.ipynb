```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```
df = pd.read_csv('/content/sample_data/high_diamond_ranked_10min.csv')
```

```
df.head()
```

```
columns = ['redGoldDiff', 'redExperienceDiff', ]
```

```
df = df.drop(columns, axis=1)
```

```
df.head()
```

Создадим новые фичи

```
df['wardpdiff'] = df.blueWardsPlaced - df.redWardsPlaced
df['dragonsdiff'] = df.blueDragons - df.redDragons
df['heraldsdiff'] = df.blueHeralds - df.redHeralds
```

```
X = df[['wardpdiff', 'dragonsdiff', 'heraldsdiff', 'blueExperienceDiff', 'blueGoldDiff']]
y = df.blueWins
```

```
X
```

```
scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```
y.head()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state = 10)
```

Логистическая регрессия


```
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {}'.format(logreg.score(X_test, y_test)))
```

Матрица ошибок



```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```
